## Функционал

В качестве новых фич вы можете реализовать:

- Изменение скорости шарика при столкновении с движущейся ракеткой. *Замечание.* Здесь имеется в виду придание дополнительной скорости мячику, если ракетка движется в том же направлении, что и мячик, либо уменьшение скорости, если в разном.

- Изменение скорости и направления мячика при ударе об угол ракетки.

- Изменение размера шарика. *Замечание.* Реализация, к примеру, в меню.

- Изменение скорости и/или размера можно сделать случайным или по какому-то интересному правилу.

- Вышеперечисленное можно реализовать в виде отдельных режимов игры.
